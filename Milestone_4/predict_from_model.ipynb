```python
import joblib
import pandas as pd
import os 


MODEL_DIR = "milestone4_outputs_advanced/models"


REG_PATH = os.path.join(MODEL_DIR, "chosen_regressor.pkl")
CLF_PATH = os.path.join(MODEL_DIR, "chosen_classifier.pkl")
LE_PATH = os.path.join(MODEL_DIR, "label_encoder_lighting.pkl")

try:
    rf_reg = joblib.load(REG_PATH)
    rf_clf = joblib.load(CLF_PATH)
    le_lighting = joblib.load(LE_PATH)
except FileNotFoundError as e:
    
    print(f"âŒ Ø®Ø·Ø£: Ù„Ù… ÙŠØªÙ… Ø§Ù„Ø¹Ø«ÙˆØ± Ø¹Ù„Ù‰ Ø§Ù„Ù…Ù„Ù. Ø§Ù„Ù…Ø³Ø§Ø± Ø§Ù„Ø®Ø§Ø·Ø¦: {e.filename}")
    print(f"ÙŠØ±Ø¬Ù‰ Ø§Ù„ØªØ£ÙƒØ¯ Ù…Ù† Ø£Ù† Ù…Ø¬Ù„Ø¯ '{MODEL_DIR}' Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ø¬Ù„Ø¯ Ø§Ù„ØªØ´ØºÙŠÙ„ Ø§Ù„Ø­Ø§Ù„ÙŠ.")
   
    raise 


new_data = {
    "vehicle_count": 25,
    "vehicle_speed": 22.5,
    "solar_energy_level": 0.6,
    "hour": 18,
    "weekday": 2,
    "is_weekend": 0,
    "veh_roll_3": 24,
    "veh_count_lag_1": 28,
    "veh_count_lag_2": 30,
    "veh_count_lag_3": 25,
    "street_name_enc": 1,
    "light_level_enc": 0,
    "weather_enc": 0,
    "traffic_light_enc": 1
}

X_new = pd.DataFrame([new_data])

pred_count = rf_reg.predict(X_new)[0]

pred_light_encoded = rf_clf.predict(X_new).astype(int) 
pred_light = le_lighting.inverse_transform(pred_light_encoded)[0]

print("ğŸš— Expected vehicle count next:", round(pred_count, 2))
print("ğŸ’¡ Predicted lighting demand:", pred_light)
```
